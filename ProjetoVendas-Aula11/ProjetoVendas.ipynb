```python
vendas_agosto = {
    "04/08/2025": 2350,
    "05/08/2025": 1980,
    "06/08/2025": 2540,
    "07/08/2025": 2100,
    "08/08/2025": 3050,
    "11/08/2025": 2650,
    "12/08/2025": 2180,
    "13/08/2025": 2760,
    "14/08/2025": 2920,
    "15/08/2025": 3100,
    "18/08/2025": 2300,
    "19/08/2025": 2430,
    "20/08/2025": 2210,
    "21/08/2025": 2790,
    "22/08/2025": 2970,
    "25/08/2025": 3100,
    "26/08/2025": 2880,
    "27/08/2025": 2600,
    "28/08/2025": 2500,
    "29/08/2025": 3400
}

print('-- Relatório de vendas do mês de agosto --\n'.center(100))

valores_vendas = list(vendas_agosto.values())
lista_dias = list(vendas_agosto.keys())

total_vendas_mes = sum(valores_vendas)

print(f'Total do mês: R$ {total_vendas_mes}')

media_vendas = total_vendas_mes / len(vendas_agosto)

print(f'Média de vendas do mês: R$ {media_vendas}')

semana_1 = {}
semana_2 = {}
semana_3 = {}
semana_4 = {}

for i in range(0, 5):
    semana_1[lista_dias[i]] = valores_vendas[i]
    semana_2[lista_dias[i+5]] = valores_vendas[i+5]
    semana_3[lista_dias[i+10]] = valores_vendas[i+10]
    semana_4[lista_dias[i+15]] = valores_vendas[i+15]

total_semana_1 = sum(list(semana_1.values()))
total_semana_2 = sum(list(semana_2.values()))
total_semana_3 = sum(list(semana_3.values()))
total_semana_4 = sum(list(semana_4.values()))

print('\n' + '-' * 100 + '\n')

print('Total por semana:')
print(f'Semana 1: R$ {total_semana_1}')
print(f'Semana 2: R$ {total_semana_2}')
print(f'Semana 3: R$ {total_semana_3}')
print(f'Semana 4: R$ {total_semana_4}')

media_semana_1 = total_semana_1 / len(semana_1)
media_semana_2 = total_semana_2 / len(semana_2)
media_semana_3 = total_semana_3 / len(semana_3)
media_semana_4 = total_semana_4 / len(semana_4)

print('\n' + '-' * 100 + '\n')

print('Média por semana:')
print(f'Semana 1: R$ {media_semana_1}')
print(f'Semana 2: R$ {media_semana_2}')
print(f'Semana 3: R$ {media_semana_3}')
print(f'Semana 4: R$ {media_semana_4}')

print('\n' + '-' * 100 + '\n')

maior_venda = max(valores_vendas)
indice_maior_venda = valores_vendas.index(maior_venda)
dia_maior_venda = lista_dias[indice_maior_venda]

print(f'Dia com maior valor de venda: {dia_maior_venda} com o valor R$ {maior_venda}')

print('\n' + '-' * 100 + '\n')

menor_venda = min(valores_vendas)
indice_menor_venda = valores_vendas.index(menor_venda)
dia_menor_venda = lista_dias[indice_menor_venda]

print(f'Dia com menor valor de venda: {dia_menor_venda} com o valor R$ {menor_venda}')

print('\n' + '-' * 100 + '\n')

diferenca_percentual = ((maior_venda - media_vendas) / media_vendas) * 100

print(f'Pico mensal: {diferenca_percentual:.2f}%')

print('\n' + '-' * 100 + '\n')
print(' -- Fim do Relatório -- '.center(100))
```
